````python
import pandas as pd
import numpy as np

def run_sales_simulation(products, num_orders):
    """
    Simulates sales for given products over a specified number of orders.

    :param products: A list of dictionaries containing product info (name, price, cost).
    :param num_orders: An integer specifying the number of orders to simulate.
    :return: A DataFrame containing the simulation results.
    """
    
    # Prepare list to hold order data
    orders = []
    
    for _ in range(num_orders):
        # Select a random product
        product = np.random.choice(products)
        
        # Generate a random quantity ordered
        quantity_ordered = np.random.randint(1, 6)  # Random number between 1 and 5
        price_each = product['price']
        cost_price = product['cost']
        
        # Calculate turnover and margin
        turnover = price_each * quantity_ordered
        margin = (price_each - cost_price) * quantity_ordered
        
        # Append order data to the list
        orders.append({
            "Order ID": len(orders) + 1,
            "Product": product['name'],
            "Quantity Ordered": quantity_ordered,
            "Price Each": price_each,
            "Cost Price": cost_price,
            "Turnover": turnover,
            "Margin": margin
        })
    
    # Create a DataFrame from the orders
    orders_df = pd.DataFrame(orders)
    
    # Calculate total sales and total margin
    total_sales = orders_df['Turnover'].sum()
    total_margin = orders_df['Margin'].sum()
    
    print(orders_df)
    print(f"Total Sales: ${total_sales:.2f}")
    print(f"Total Margin: ${total_margin:.2f}")
    
    return orders_df

# Example products to simulate
products = [
    {"name": "iPhone", "price": 700.00, "cost": 231.00},
    {"name": "Lightning Charging Cable", "price": 14.95, "cost": 7.475},
    {"name": "Wired Headphones", "price": 11.99, "cost": 5.995},
    {"name": "27in FHD Monitor", "price": 149.99, "cost": 97.4935}
]

# Running the simulation
num_orders = 10  # Specify the number of orders to simulate
run_sales_simulation(products, num_orders)
````

````python


import pandas as pd
import numpy as np

def run_sales_simulation_by_supplier(products, num_orders):
    """
    Simulates sales for products grouped by supplier over a specified number of orders.

    :param products: A list of dictionaries containing product info (name, price, cost, supplier).
    :param num_orders: An integer specifying the number of orders to simulate.
    :return: A summary DataFrame with total sales and total margin per supplier.
    """
    
    # Prepare a dictionary to hold results by supplier
    supplier_results = {}
    
    for _ in range(num_orders):
        # Select a random product
        product = np.random.choice(products)
        
        # Generate a random quantity ordered
        quantity_ordered = np.random.randint(1, 6)  # Random number between 1 and 5
        price_each = product['price']
        cost_price = product['cost']
        
        # Calculate turnover and margin
        turnover = price_each * quantity_ordered
        margin = (price_each - cost_price) * quantity_ordered
        
        # Get the supplier
        supplier = product['supplier']

        # Add results to the supplier's total in the dictionary
        if supplier not in supplier_results:
            supplier_results[supplier] = {
                'Total Sales': 0,
                'Total Margin': 0
            }
        
        supplier_results[supplier]['Total Sales'] += turnover
        supplier_results[supplier]['Total Margin'] += margin
    
    # Convert results to DataFrame
    summary_df = pd.DataFrame.from_dict(supplier_results, orient='index').reset_index()
    summary_df.columns = ['Supplier', 'Total Sales', 'Total Margin']
    
    print(summary_df)
    return summary_df

# Example products with suppliers
products = [
    {"name": "iPhone", "price": 700.00, "cost": 231.00, "supplier": "Apple"},
    {"name": "Lightning Charging Cable", "price": 14.95, "cost": 7.475, "supplier": "Apple"},
    {"name": "Wired Headphones", "price": 11.99, "cost": 5.995, "supplier": "Apple"},
    {"name": "27in FHD Monitor", "price": 149.99, "cost": 97.4935, "supplier": "Dell"},
    {"name": "Gaming Mouse", "price": 39.99, "cost": 20.00, "supplier": "Logitech"},
    {"name": "Mechanical Keyboard", "price": 89.99, "cost": 50.00, "supplier": "Logitech"}
]

# Running the simulation
num_orders = 100  # Specify the number of orders to simulate
run_sales_simulation_by_supplier(products, num_orders)
````
